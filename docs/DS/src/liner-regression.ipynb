```python
import matplotlib.pyplot as plt


x = list(range(100))

def y_func(x,m,c):
    y = []
    for each_x in x:
        y.append(m*each_x+c) 
    return y
y = y_func(x,1,1)
print(y)
plt.plot(x,y)
```

```python
y = y_func(x,1,2)
print(y)
plt.plot(x,y)
```
