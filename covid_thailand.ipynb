```python
df1 = pd.read_csv("covid_dataset_th1.csv", encoding="utf-8-sig")
df2 = pd.read_csv("covid_dataset_th2.csv", encoding="utf-8-sig")
```

```python
df = pd.concat([df1, df2])
```

```python
df.dtypes
```

```python
df['year_ef'] = df['EFFECTIVE_DATE'].dt.year
```

```python
date_col = ['announce_date', 'Notified date']
for col in date_col:
    df[col] = pd.to_datetime(df[col])
```
